# Calculate Formality using F-Score

From Formality of Language: definition, measurement, and behavioral determinants

F = (noun freq + adj freq + prep freq + article freq - pronoun freq - verb freq - adverb freq - interjection freq + 100)/2

```python
def calcFscore(tagged_sents):
    word_count = 0
    noun_count = 0
    adj_count = 0
    prep_count = 0
    art_count = 0
    verb_count = 0
    pro_count = 0
    adv_count = 0
    inter_count = 0
    for sent in tagged_sents:
        word_count = word_count + len(sent)
        for word, tag in sent:
            if tag.startswith('N'):
                noun_count +=1
            elif tag.startswith('J'):
                adj_count +=1
            elif tag.startswith('I'):
                prep_count +=1
            elif tag.startswith('A'):
                art_count +=1
            elif tag.startswith('V'):
                verb_count +=1
            elif tag.startswith('P'):
                pro_count +=1
            elif tag.startswith('R'):
                adv_count +=1
            elif tag.startswith('U'):
                inter_count +=1
    if word_count == 0:
        return np.nan
    noun_freq = noun_count/word_count
    adj_freq = adj_count/word_count
    prep_freq = prep_count/word_count
    art_freq = art_count/word_count
    verb_freq = verb_count/word_count
    pro_freq = pro_count/word_count
    adv_freq = adv_count/word_count
    inter_freq = inter_count/word_count
    return (noun_freq + adj_freq+ prep_freq + art_freq - verb_freq - pro_freq - adv_freq - inter_freq + 100)/2
```

```python
data['f_score'] = data.ngram_tags.map(calcFscore)
data.f_score
```

There is not a lot of variation in formality scores across the job descriptions. As an alternative, we can turn to one dimension of Biber's dimensions of English- persuasivenes. He suggest that the following features indicate 'overt argumentation and persuasion':

* modals (prediction, necessity, possibilitity)
* suasive verbs
* conditional subordination
* infininitive verbs

Other features also seem promising for indicating a desire to make a candidate think about the possibilities of working for the company:

* question marks (often use to ask "are you passionate about...?")
* exclamation points ("come join us!")
* description length


```python
#http://en.wikipedia.org/wiki/English_modal_verbs
modals = ['can', 'could', 'may', 'might', 'will', 'would', 'should', 'shall']
suasive_verbs = ['insist', 'demand', 'desire']
#http://www1.icsi.berkeley.edu/~kay/bcg/lec07.html
conditional_subordination = ['if', 'unless', 'otherwise']
```

```python
def getModals(word_lists):
    modal_count = 0
    for word_list in word_lists:
        for word in word_list:
            if word.lower() in modals:
                modal_count +=1
    return modal_count
```

```python
def getModalsNormalized(word_lists):
    modal_count = 0
    word_count = 0
    for word_list in word_lists:
        for word in word_list:
            if word.lower() in modals:
                modal_count +=1
            word_count +=1
    if word_count == 0:
        return np.nan
    else:
        return modal_count/word_count
```

```python
data['modals'] = data.words.map(getModals)
data.modals.describe()
```

```python
data['modals_normalized'] = data.words.map(getModalsNormalized)
data.modals_normalized.describe()
```

```python
def getInfinitives(tagged_sents):
    infinitive_count = 0
    for sent in tagged_sents:
        for word, tag in sent:
            if tag == 'TO':
                infinitive_count +=1
    return infinitive_count
```

```python
data['infinitives'] = data.ngram_tags.map(getInfinitives)
data.infinitives.describe()
```

```python
def getInfinitivesNormalized(tagged_sents):
    infinitive_count = 0
    word_count = 0
    for sent in tagged_sents:
        for word, tag in sent:
            if tag == 'TO':
                infinitive_count +=1
            word_count +=1
    if word_count == 0:
        return np.nan
    else:
        return infinitive_count/word_count
```

```python
data['infinitives_normalized'] = data.ngram_tags.map(getInfinitivesNormalized)
data.infinitives_normalized.describe()
```

```python
def getConditionals(word_lists):
    conditional_count = 0
    for sent in word_lists:
        for word in sent:
            if word.lower() in conditional_subordination:
                conditional_count +=1
    return conditional_count
```

```python
data['conditionals'] = data.words.map(getConditionals)
data.conditionals.describe()
```

```python
def getConditionalsNormalized(word_lists):
    conditional_count = 0
    word_count = 0
    for sent in word_lists:
        for word in sent:
            if word.lower() in conditional_subordination:
                conditional_count +=1
            word_count +=1
    if word_count == 0:
        return np.nan
    else:
        return conditional_count/word_count
```

```python
data['conditionals_normalized'] = data.words.map(getConditionalsNormalized)
data.conditionals_normalized.describe()
```

```python
puncts = ['!','?']
def getPunctuation(word_lists):
    punct_count = 0
    for sent in word_lists:
        for word in sent:
            if word in puncts:
                punct_count +=1
    return punct_count
```

```python
data['punctuation'] = data.words.map(getPunctuation)
data.punctuation.describe()
```

```python
puncts = ['!','?']
def getPunctuationNormalized(word_lists):
    punct_count = 0
    word_count = 0
    for sent in word_lists:
        for word in sent:
            if word in puncts:
                punct_count +=1
            word_count +=1
    if word_count == 0:
        return np.nan
    else:
        return punct_count/word_count
```

```python
data['punctuation_normalized'] = data.words.map(getPunctuationNormalized)
data.punctuation_normalized.describe()
```
