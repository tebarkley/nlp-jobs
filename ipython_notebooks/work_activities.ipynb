## Extract work activities- chunking

Get all sentences that begin with a verb, as these are almost always indicative of work activities. 

```python
imperatives = []
for pos_sents in data.ngram_tags.tolist():
    for pos_sent in pos_sents:
        if pos_sent[0][1].startswith('V') and len(pos_sent) > 6:
            imperatives.append(pos_sent)
```

Do some filtering on the imperatives based on things that we want to exclude:

* sentences that contain 'we'- as these are more about the company than the activities
* sentences that contain 'you'- as these are more about interactions with the company that work activities
* sentences that end in exclamation points
* sentences that begin with founded, based, started, or located as these are about the company
* sentences that begin with please, as these are instructions for applying

```python
len(imperative_sents)
```

```python
imperatives_keep = []
imperatives_keep = [sent for sent in imperatives if 'we' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives if 'us' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if 'you' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if 'founded' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if 'based' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if 'started' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if 'located' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if '!' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if 'please' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if 'resume' not in [word.lower() for word, pos in sent]]
imperatives_keep = [sent for sent in imperatives_keep if 'resumes' not in [word.lower() for word, pos in sent]]
len(imperatives_keep)
```

```python
imperative_sents = []
for sent in imperatives_keep:
    imperative_sents.append(' '.join([word for word, tag in sent]))
```

From these sentences containing work activities, we now want to to use chunking to extract the specific activities

```python
def getChunks(sentences, tag=tag, grammar=grammar):
    cp = nltk.RegexpParser(grammar)
    chunks = []
    for sent in sentences:
        tree = cp.parse(sent)
        for subtree in tree.subtrees():
            if subtree.node == tag: 
                chunks.append(subtree.leaves())
    return chunks
```

Extract basic verb, optional adjective, noun phrase, plus an optional second noun phrase

```python
tag = 'VP1'
grammar = r"""
  VP1: {<VB|VBG.*|VBZ.*><JJ|VBG>?<N.*>+(<CC.*><JJ|VBG>?<N.*>)?}   
"""
vp1 = getChunks(imperatives_keep, tag, grammar)
print len(vp1)
vp1
```

Extract 'verb, verb, and verb phrases'. Can consider changing the POS of the verbs in the chain, as many have been incorrectly tagged as nouns.

```python
tag='VP3'
grammar = r"""
  VP3: {<VB|VBG.*|VBZ.*><TO|IN>?(<,><VB|VBG.*|VBZ.*|N.*><TO|IN>?<,>?)*<CC.*><VB|VBG.*|VBZ.*|N.*><TO|IN>?<JJ.*>?<N.*>+}   # one of more adjectives or nouns, followed by a noun or cardinal number
"""
vp3 = getChunks(imperatives_keep, tag, grammar)
print len(vp3)
vp3
```

## Fixing POS tags

Before lemmatizing the nouns, verbs, and adjectives, we can fix the POS tagging of chained verbs and nouns, by assuming that the POS of the word following a '<verb> and' is also a verb

```python
vp3_posfix = []
for chunk in vp3:
    new_chunk = []
    prev_word2 = ''
    prev_word = ''
    for word, pos in chunk:
        if prev_word2.startswith('V') or prev_word2==',': #if two words before is a verb or a comma
            if prev_word.startswith('CC') or prev_word == ',': #if the word before is and or a comma
                pos = prev_word2
        new_chunk.append((word, pos))
        prev_word2 = prev_word
        prev_word = pos
    vp3_posfix.append(new_chunk)
```

## Frequency Distributions

First, put all of the work activities together in the same list

```python
all_work_activities = []
[all_work_activities.append(act) for act in vp1]
[all_work_activities.append(act) for act in vp3_posfix]
[all_work_activities.append(act) for act in vp4]
[all_work_activities.append(act) for act in vp5]
[all_work_activities.append(act) for act in vp6]
[all_work_activities.append(act) for act in vp7]
len(all_work_activities)
```

```python
all_work_activities_lemmatized = []
work_activity_nouns = []
work_activity_verbs = []
work_activity_adjs = []
for act in all_work_activities:
    word_list = []
    for word, tag in act:
        if tag.startswith('N') and not tag.startswith('NP'):
            if wn.morphy(word.lower(), wn.NOUN): # if the lemma exists in wordnet
#                 word = wn.morphy(word.lower(), wn.NOUN)
                work_activity_nouns.append(wn.morphy(word.lower(), wn.NOUN))  
        elif tag.startswith('V'):
            if wn.morphy(word.lower(), wn.VERB): # if the lemma exists in wordnet
                word = wn.morphy(word.lower(), wn.VERB)
                work_activity_verbs.append(word)  
        elif tag.startswith('J'):
            if wn.morphy(word.lower(), wn.ADJ): # if the lemma exists in wordnet
#                 word = wn.morphy(word.lower(), wn.ADJ)
                work_activity_adjs.append(wn.morphy(word.lower(), wn.ADJ))  
        word_list.append(word.lower())
    all_work_activities_lemmatized.append(word_list)
```

```python
nouns_fd = nltk.FreqDist(work_activity_nouns)
nouns_fd.items()[:100]
```

```python
verbs_fd = nltk.FreqDist(work_activity_verbs)
verbs_fd.items()[:100]
```

```python
adjs_fd = nltk.FreqDist(work_activity_adjs)
adjs_fd.items()[:100]
```

We can also check out the most common bigrams and trigrams

```python
stop_words = nltk.corpus.stopwords.words('english')
import string
punctuation = string.punctuation
to_remove = stop_words
[to_remove.append(punc) for punc in punctuation]
all_bigrams = [nltk.bigrams(act) for act in all_work_activities_lemmatized]
all_bigrams_list = []
for bigrams in all_bigrams:
    [all_bigrams_list.append(bigram) for bigram in bigrams if bigram[0] not in to_remove and bigram[1] not in to_remove]
bigrams_fd = nltk.FreqDist(all_bigrams_list)
bigrams_fd.items()[:100]
```

Do the same thing with trigrams

```python
all_trigrams = [nltk.trigrams(act) for act in all_work_activities_lemmatized]
all_trigrams_list = []
for trigrams in all_trigrams:
    [all_trigrams_list.append(trigram) for trigram in trigrams if trigram[0] not in to_remove and trigram[1] not in to_remove and trigram[2] not in to_remove]
trigrams_fd = nltk.FreqDist(all_trigrams_list)
trigrams_fd.items()[:100]
```
