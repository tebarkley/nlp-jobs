## Fix POS Tagging at the beginning of a word

A lot of the first words in a sentence are verbs, but have been tagged as nouns. Getting good work activities relies on these being tagged correctly. We can use the following rules:

* if the first word in a sentence is tagged as a noun and is followed by a word of POS tag <IN><AT><RB><ADJ>, make it a verb
* otherwise, if it has more verb than noun synsets in wordnet, tag it as a verb

```python
def fix_tags(sents):
    fixed_tags = []
    for sent in sents:
        first_word = sent[0][0]
        first_tag = sent[0][1]
        if first_tag.startswith('N') and len(sent)>1:
            next_tag = sent[1][1]
            new_sent = []
            if len(wn.synsets(first_word.lower(), pos=wn.VERB)) > len(wn.synsets(first_word.lower(), pos=wn.NOUN)):
                first_tag = 'VB'
            
#             if next_tag.startswith('J') or next_tag.startswith('A') or next_tag.startswith('R') or next_tag.startswith('I'):
#                 first_tag = 'VB'
#             else:
#                 if len(wn.synsets(first_word.lower(), pos=wn.VERB)) >= len(wn.synsets(first_word.lower(), pos=wn.NOUN)):
#                     first_tag = 'VB'
            new_sent.append((first_word, first_tag))
            [new_sent.append((word,tag)) for word, tag in sent[1:]] 
            fixed_tags.append(new_sent)
        else:
            fixed_tags.append(sent)
    return fixed_tags
```

```python
data['ngram_tags'] = data.ngram_tags.map(fix_tags)
```

From these sentences containing work activities, we now want to to use chunking to extract the specific activities

```python
tag=''
grammar=''
def getChunks(sentences, tag=tag, grammar=grammar):
    cp = nltk.RegexpParser(grammar)
    chunks = []
    for sent in sentences:
        tree = cp.parse(sent)
        for subtree in tree.subtrees():
            if subtree.node == tag: 
                chunks.append(subtree.leaves())
    return chunks
```

## Fixing POS tags

Before lemmatizing the nouns, verbs, and adjectives, we can fix the POS tagging of chained verbs and nouns, by assuming that the POS of the word following a '<verb> and' is also a verb

```python
vp3_posfix = []
for chunk in vp3:
    new_chunk = []
    prev_word2 = ''
    prev_word = ''
    for word, pos in chunk:
        if prev_word2.startswith('V') or prev_word2==',': #if two words before is a verb or a comma
            if prev_word.startswith('CC') or prev_word == ',': #if the word before is and or a comma
                pos = prev_word2
        new_chunk.append((word, pos))
        prev_word2 = prev_word
        prev_word = pos
    vp3_posfix.append(new_chunk)
```

Write a function that applies to the ngram_tags column that extracts each verb phrase from each job description and adds that into a column

```python
def keepImperative(sent):
    if 'we' in [word.lower() for word, pos in sent]:
        return False
    elif 'us' in [word.lower() for word, pos in sent]:
        return False
    elif 'you' in [word.lower() for word, pos in sent]:
        return False
    elif 'founded' in [word.lower() for word, pos in sent]:
        return False
    elif 'based' in [word.lower() for word, pos in sent]:
        return False
    elif 'located' in [word.lower() for word, pos in sent]:
        return False
    elif 'started' in [word.lower() for word, pos in sent]:
        return False
    elif '!' in [word.lower() for word, pos in sent]:
        return False
    elif 'please' in [word.lower() for word, pos in sent]:
        return False
    elif 'resume' in [word.lower() for word, pos in sent]:
        return False
    elif 'resumes' in [word.lower() for word, pos in sent]:
        return False
    else:
        return True
```

```python
def getImperatives(tagged_sents):
    imperatives = []
    for sent in tagged_sents:
        if sent[0][1].startswith('V') and len(sent) > 6:
            if keepImperative(sent):
                imperatives.append(sent)
    return imperatives
```

```python
vp3_posfix = []
for chunk in vp3:
    new_chunk = []
    prev_word2 = ''
    prev_word = ''
    for word, pos in chunk:
        if prev_word2.startswith('V') or prev_word2==',': #if two words before is a verb or a comma
            if prev_word.startswith('CC') or prev_word == ',': #if the word before is and or a comma
                pos = prev_word2
        new_chunk.append((word, pos))
        prev_word2 = prev_word
        prev_word = pos
    vp3_posfix.append(new_chunk)
```

```python
def chainedVerbsTagFix(chunks):
    pos_fix = []
    for chunk in chunks:
        new_chunk = []
        prev_word2 = ''
        prev_word = ''
        for word, pos in chunk:
            if prev_word2.startswith('V') or prev_word2==',': #if two words before is a verb or a comma
                if prev_word.startswith('CC') or prev_word == ',': #if the word before is and or a comma
                    pos = prev_word2
            new_chunk.append((word, pos))
            prev_word2 = prev_word
            prev_word = pos
        pos_fix.append(new_chunk)
    return pos_fix
```

```python
def extractWorkActivities(sents):
    work_activities = []
    tag = 'VP1'
    grammar = r"""
      VP1: {<VB|VBG.*|VBZ.*><JJ|VBG>?<N.*>+(<CC.*><JJ|VBG>?<N.*>)?}   
    """
    [work_activities.append(chunk) for chunk in getChunks(sents, tag, grammar)]
    tag = 'VP4'
    grammar = r"""
      VP4: {<VB|VBG.*|VBZ.*><RB>?<IN><JJ|VBG.*>?<N.*>+<VB.*>?}   # one of more adjectives or nouns, followed by a noun or cardinal number
    """
    [work_activities.append(chunk) for chunk in getChunks(sents, tag, grammar)]
    tag = 'VP5'
    grammar = r"""
      VP5: {<VB|VBG.*|VBZ.*><JJ|VBG>?<N.*>+(<,><JJ|VBG>?<N.*><,>?)*<CC.*><JJ|VBG>?<N.*>}   # one of more adjectives or nouns, followed by a noun or cardinal number
    """
    [work_activities.append(chunk) for chunk in getChunks(sents, tag, grammar)]
    tag = 'VP6'
    grammar = r"""
      VP6: {<VB|VBG.*|VBZ.*><TO><JJ|VBG>?<N.*>+}   # adhere to noun
    """
    [work_activities.append(chunk) for chunk in getChunks(sents, tag, grammar)]
    tag = 'VP7'
    grammar = r"""
      VP7: {<VB|VBG.*|VBZ.*><AT><JJ|VBG>*<N.*>+(<IN><JJ|VBG>*<N.*>+)?}   # adhere to noun
    """
    [work_activities.append(chunk) for chunk in getChunks(sents, tag, grammar)]
    tag='VP3'
    grammar = r"""
      VP3: {<VB|VBG.*|VBZ.*><TO|IN>?(<,><VB|VBG.*|VBZ.*|N.*><TO|IN>?<,>?)*<CC.*><VB|VBG.*|VBZ.*|N.*><TO|IN>?<JJ.*>?<N.*>+}   # one of more adjectives or nouns, followed by a noun or cardinal number
    """
    chained_verbs = getChunks(sents, tag, grammar)
    [work_activities.append(chunk) for chunk in chainedVerbsTagFix(chained_verbs)]
    return work_activities
    
```

```python
def getWorkActivities(tagged_sents):
    #start by getting imperatives
    imperatives = getImperatives(tagged_sents)
    work_activities = extractWorkActivities(imperatives)
    return work_activities
```

Next, add a column that lemmatizes the verbs in the work activities

```python
def lemmatizeActivities(activities):
    lemmatized_activities = []
    for act in activities:
        word_list = []
        for word, tag in act:
            if tag.startswith('V'):
                if wn.morphy(word.lower(), wn.VERB): # if the lemma exists in wordnet
                    word = wn.morphy(word.lower(), wn.VERB)
            word_list.append(word.lower())
        lemmatized_activities.append(word_list)
    return lemmatized_activities    
```

Now add columns containing lists of lemmatized verbs, nouns, and adjectives, for use in frequency distributions

```python
def getNouns(work_activities):
    nouns = []
    for act in work_activities:
        for word, tag in act:
            if tag.startswith('N'):
                if wn.morphy(word.lower(), wn.NOUN): # if the lemma exists in wordnet
                    word = wn.morphy(word.lower(), wn.NOUN)
                nouns.append(word)
    return nouns

def getAdjs(work_activities):
    adjs = []
    for act in work_activities:
        for word, tag in act:
            if tag.startswith('J'):
                if wn.morphy(word.lower(), wn.ADJ): # if the lemma exists in wordnet
                    word = wn.morphy(word.lower(), wn.ADJ)
                adjs.append(word)
    return adjs

def getVerbs(work_activities):
    verbs = []
    for act in work_activities:
        for word, tag in act:
            if tag.startswith('V'):
                if wn.morphy(word.lower(), wn.VERB): # if the lemma exists in wordnet
                    word = wn.morphy(word.lower(), wn.VERB)
                verbs.append(word)
    return verbs
```

```python
data['nouns'] = data.work_activities.map(getNouns)
data['adjs'] = data.work_activities.map(getAdjs)
data['verbs'] = data.work_activities.map(getVerbs)
```

```python
stop_words = nltk.corpus.stopwords.words('english')
import string
punctuation = string.punctuation
to_remove = stop_words
[to_remove.append(punc) for punc in punctuation]

def getBigrams(lemma_activity_list):
    all_bigrams = [nltk.bigrams(act) for act in lemma_activity_list]
    all_bigrams_list = []
    for bigrams in all_bigrams:
        [all_bigrams_list.append(bigram) for bigram in bigrams if bigram[0] not in to_remove and bigram[1] not in to_remove]
    return all_bigrams_list

def getTrigrams(lemma_activity_list):
    all_trigrams = [nltk.trigrams(act) for act in lemma_activity_list]
    all_trigrams_list = []
    for trigrams in all_trigrams:
        [all_trigrams_list.append(trigram) for trigram in trigrams if trigram[0] not in to_remove and trigram[1] not in to_remove and trigram[2] not in to_remove]
    return all_trigrams_list
```
